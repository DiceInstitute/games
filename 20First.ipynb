```python
#Let's play a game
#Rule is simple
#Whoever reaches 20 first will be the winner
#You can add +1 or +2
#Starting point will be 0
#+1 or +2 in every step is a must
```

```python
# Do a toss randomly for who will start the game.
# If the output comes 1 then computer will be first
# If the output comes 0 then you'll be the first to enter the number

# Imported the library for the random function
import random

# Imported the library for the time management
import time

# Initializing the next turn
nextTurn = 'c'

# Initilizing your and computer turn variables
computerTurn = 0
yourTurn = 0

# Initializng the sum
sum20 = 0

# Initializing step count
steps = 0

# Assigned a variable to get the random integer value among 0 and 1
choice = random.randint(0, 1)

# Taking a decision who'll start first 0=>You , 1=>Computer
if (choice):
    print('Output came as', choice, 'So Computer won the toss.')
    print("Computer is thinking a number for 3 seconds.")

    # Delaying the process for 3 seconds
    time.sleep(3)

    # Identfying the first number randomly and assigning to the variable
    computerFirst = random.randint(1, 2)

    # Increasing step count
    steps += 1

    print("Step==>", steps, "Computer Said==>", computerFirst)

    # Increasing the turn number for computer
    nextTurn = 'y'

    # Calculating the total Sum
    sum20 += computerFirst
else:
    print('Output came as', choice, 'So You won the toss')
    print("Please enter your first number after thinking 3 seconds")

    # Delaying the process for 3 seconds
    time.sleep(3)

    # Taking the first number from you and assiging to the variable
    yourFirst = int(input("Your first number is: "))
    # Increasing step count
    steps += 1

    print("Step==>", steps, "You Said==>", yourFirst)

    # Calculating the total Sum
    sum20 += yourFirst

while sum20 < 20:
    if (nextTurn == 'c'):
        if (sum20 >= 18):
            computerTurn = 20 - sum20
            sum20 = 20
            steps += 1
            print("Step==>", steps, "Computer Said==>", sum20)
            print("Sorry you've lost, computer won this game, better luck next time!")
        else:
            print("Computer is thinking a number for 3 seconds.")
            # Delaying the process for 3 seconds
            time.sleep(3)

            # Identfying the number randomly and assigning to the variable
            computerTurn = random.randint(1, 2)

            # Increasing step count
            steps += 1

            # Calculating the total Sum
            sum20 += computerTurn

            print("Step==>", steps, "Computer Said==>", sum20)

            # Increasing the turn number for computer
            nextTurn = 'y'
    else:
        if (sum20 >= 18):
            yourTurn = 20 - sum20
            sum20 = 20
            steps += 1
            print("Step==>", steps, "You Said==>", sum20)
            print("Hoorray you've won the game, congratulations for defeating computer!")
        else:
            print("You take time to think a number for 3 seconds.")
            # Delaying the process for 3 seconds
            time.sleep(3)

            # Identfying the number randomly and assigning to the variable
            yourTurn = int(input("Your number is: "))

            # Increasing step count
            steps += 1

            # Calculating the total Sum
            sum20 += yourTurn

            print("Step==>", steps, "You Said==>", sum20)

            # Increasing the turn number for computer
            nextTurn = 'c'
```
