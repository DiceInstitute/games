```python
#Let's play a game
#Rule is simple
#Whoever reaches 20 as sum first will be the winner
#You can add +1 to +6
#Starting point will be 0
#+1 to +6 in every step is a must
```

```python
# Do a toss randomly for who will start the game.
# To start the summation you need to have 1
# If you get 3 at any point of time your summation will be deducted by 5 points
# If the output comes 1 then computer will be first
# If the output comes 0 then you'll be the first to enter the number

# Imported the library for the random function
import random

# Imported the library for the time management
import time

# Initializing the next turn
nextTurn = 'c'

# Initilizing your and computer turn variables
computerStart = 0
yourStart = 0

# Initializng the sum
computerSum = 0
yourSum = 0

# Initializing step count
steps = 0

# Assigned a variable to get the random integer value among 0 and 1
choice = random.randint(0, 1)

# Taking a decision who'll start first 0=>You , 1=>Computer
if (choice):
    print('Output came as', choice, 'So Computer won the toss.')
    print("Computer is rolling the die...")

    # Delaying the process for 3 seconds
    time.sleep(3)

    # Identfying the first number randomly and assigning to the variable
    computerFirst = random.randint(1, 6)

    # Assigning computerTurn if computer gets the starting value 1
    if computerFirst == 1:
        computerStart = 1
        print("Computer Started, computer is rolling it's die now.")
        time.sleep(3)
        computerSum = random.randint(1, 6)

    # Increasing step count
    steps += 1

    print("Step==>", steps, "Computer first number", computerFirst, "Computer Total is==>", computerSum)

    # Increasing the turn number for computer
    nextTurn = 'y'

else:
    print('Output came as', choice, 'So You won the toss')
    print("Your die is rolling.....")

    # Delaying the process for 3 seconds
    time.sleep(3)

    # Identfying the first number randomly and assigning to the variable
    yourFirst = random.randint(1, 6)

    # Assigning computerTurn if computer gets the starting value 1
    if yourFirst == 1:
        yourStart = 1
        print("Your Started, you are rolling your die now.")
        time.sleep()
        yourSum = random.randint(1, 6)

    # Increasing step count
    steps += 1

    print("Step==>", steps, "Your first number", yourFirst, "Your sum is==>", yourSum)
    # Increasing the turn number for computer
    nextTurn = 'c'

while yourSum < 100 and computerSum < 100:
    steps += 1
    if nextTurn == 'c':
        # Delaying the process for 3 seconds
        time.sleep(3)
        temp = random.randint(1, 6)
        if computerStart == 1:
            if temp == 3:
                print("Computer got 3 so deducting 5 points from the sum.")
                computerSum = computerSum - 5
            else:
                computerSum = computerSum + temp
        else:
            if temp == 1:
                print("Computer got 1 so computer can start now. Computer rolling die again.")
                computerStart = 1
                 # Delaying the process for 3 seconds
                time.sleep(3)
                temp = random.randint(1, 6)
                if temp == 3:
                    computerSum = computerSum - temp
                else:
                    computerSum = computerSum + temp
        if computerSum >= 100:
            print("Steps==>", steps, "Computer got", temp, "Computer sum is:", computerSum)
            print("Computer won this game. Better luck next time.")
        else:
            print("Steps==>", steps, "Computer got", temp, "Computer sum is:", computerSum)
        nextTurn = 'y'
    else:
         # Delaying the process for 3 seconds
        time.sleep(3)
        temp = random.randint(1, 6)
        if yourStart == 1:
            if temp == 3:
                print("You got 3 so deducting 5 points from your sum.")
                yourSum = yourSum - temp
            else:
                yourSum = yourSum + temp
        else:
            if temp == 1:
                print("You got 1 so you can start now. You are rolling die again.")
                yourStart = 1
                 # Delaying the process for 3 seconds
                time.sleep(3)
                temp = random.randint(1, 6)
                if yourStart == 1:
                    if temp == 3:
                        print("You got 3 so deducting 5 points from your sum.")
                        yourSum = yourSum - temp
                    else:
                        yourSum = yourSum + temp
        if yourSum >= 100:
            print("Steps==>", steps, "You got", temp, "Your sum is:", yourSum)
            print("You won this game. Congratulations.")
        else:
            print("Steps==>", steps, "You got", temp, "Your sum is:", yourSum)
        nextTurn = 'c'

```
